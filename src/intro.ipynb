You can also test compilation running with the following python script:

```python
%%writefile program.cu
/// Write a simple CUDA program to program.cu
#include <stdio.h>
#include <cuda_runtime.h>

__global__ void helloCuda() {
    printf("Hello Cuda from thread %d\\n", threadIdx.x);
}

int main() {
    printf("Starting CUDA program...\\n");
    helloCuda<<<1, 5>>>();
    cudaDeviceSynchronize();
    printf("Program completed.\\n");
    return 0;
}
```

```python
# Compile the CUDA program
!nvcc program.cu -o program_executable

# Run the compiled program
!./program_executable
```
